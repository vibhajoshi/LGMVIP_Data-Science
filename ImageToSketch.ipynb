**Convert to grey image**

```python
grey_filter=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
cv2_imshow(grey_filter)
```

**Invert the image**

```python
invert=cv2.bitwise_not(grey_filter)
cv2_imshow(invert)
```

**Blur the image**

```python
blur=cv2.GaussianBlur(invert,(21,21),0)
cv2_imshow(blur)
```

**Invert the blurred image**

```python
invertedblur=cv2.bitwise_not(blur)
cv2_imshow(invertedblur)
```

**Final sketch**

```python
sketch_filter=cv2.divide(grey_filter,invertedblur,scale=255.0)
cv2_imshow(sketch_filter)
cv2.imwrite("sketch.jpg",sketch_filter)
```
